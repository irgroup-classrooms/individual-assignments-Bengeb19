```python
import re 
from collections import Counter


def main():
    
    # Read the CSV file with the product orders
    with open('/Users/gengebhardt/Documents/Uni/Semester_3/Datenmodellierung/Übungen/orders.csv') as f_in:
        text = f_in.read()

    # Define the regular expression to extract all order numbers
    regex_orders = r'\d+'
    regex_Onumbers = r'(?<=#)\d+'
    regex_Pnames = r'(?<=Product: )[A-Za-z]+'
    regex_prices = r'(?<=\$)\d+\.\d+'
    regex_order_dates = r'(?<=Date: )\d{4}-\d{2}-\d{2}'
    regex_new_dates = r'(?<=Date: )\d{2}-\d{2}-\d{4}'
    regex_Pnames6 = r'(?<=Product: )[A-Za-z]{7,}'
    regex_order_prices99 = r'Product: [A-Za-z]+, Price: \$\d+\.99'
    
    new_dates_text = re.sub(r'(\d{4})-(\d{2})-(\d{2})', r'\3-\2-\1', text)

    # Match the regex with the text
    orders = re.findall(regex_orders, text) 
    order_numbers = re.findall(regex_Onumbers, text) #1
    product_names = re.findall(regex_Pnames, text) #2
    prices = re.findall(regex_prices, text) #3
    order_dates = re.findall(regex_order_dates, text) #4
    
    orders500 = [] #5
    prices_float =[]
    for price in prices:
        price_float = float(price)
        prices_float.append(price_float)
        if price_float > 500:
            orders500.append(price_float)
    
    neue_datum = re.findall(regex_new_dates, new_dates_text) #6
    
    pnames6 = re.findall(regex_Pnames6, text)#7
    
    product_counts = Counter(product_names)#8
    
    order_prices99 = re.findall(regex_order_prices99, text)#10
    
    cheapest_product = min(prices_float)
    #for price in price_float:
      #  if price > cheapest_product:
       #     cheapest_product = price
        
    

    
            
        

    # Print the results
    print('alle Zahlen')
    print(orders)
    print()
    print('Bestellnummern Aufgabe 1')
    print(order_numbers)
    print()
    print('Produktnamen Aufgabe 2')
    print(product_names)
    print()
    print('Preise Aufgabe 3')
    print(prices)
    print()
    print('Bestelldaten Aufgabe 4')
    print(order_dates)
    print()
    print('Bestellungen über $500 Aufgabe 5')
    print(orders500)
    print()
    print('Angepasstes Datum Aufgabe 6')
    print(neue_datum)
    print()
    print('Produktnamen 6 Buchstaben Aufgabe 7')
    print(pnames6)
    print()
    print('Produkt anzahlen Aufgabe 8')
    print(product_counts)
    print()
    print('Produkte mit .99 Preis Aufgabe 9')
    print(order_prices99)
    print()
    print('cheapest product Aufgabe 10')
    print(cheapest_product)
    
    
    

if __name__ == '__main__':
    main()
```
